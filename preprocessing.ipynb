```python
def place_name(df):
    df_country = df['country'].str.lower()
    df_city = df['place_slug']

    i = 0
    for c in df_country:
        df_country[i] = '-' + c
        #print(df_country[i])
        if ' ' in c:
            #print(c)
            df_country[i] = df_country[i].replace(' ','-')
        i+=1
    #print(df_country)
    
    j =0
    for cn in df_country:
        df_city[j] = df_city[j].replace(cn,'')
        #print(df_city[j])
        j+=1
    
    df['place_slug'] = df_city
    return df
```
